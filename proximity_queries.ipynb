```python
import nltk
nltk.download('punkt')
import csv
import re
from tkinter import *
```

```python
#function to process positional queries

def posqueryProcessor():
    #getting query
    query=block1.get()
    queryTok=nltk.word_tokenize(query)
    #removing special characters
    
    removetable=str.maketrans("", "", "'!@#$%^&*()_=-\|][:';:,<.>/?`~")
    queryTok=[x.translate(removetable) for x in queryTok]
    
    queryTok=[element.lower() for element in queryTok]
    
    ans={}
    res={}
    #if query is a single word
    if len(queryTok) == 1:
        try:
            ans=dict[queryTok[0]]
        except KeyError:
            print("Key Error")
                    
        if not ans:
            print ("no match")

        else:
            for x in range(1,len(ans)+1):
                if(ans[x]):
                    res.update({x:ans[x]})
                            
            block2.delete(0,END)
            ls55=list(res.keys())
            block2.insert(15,ls55)

                
     #if it is a proximity query           
    else:
        for i in range(len(queryTok)):
            try:
                ans=dict[queryTok[i]]
            except KeyError:
                print("Key Error")
    
            
            res[i]={}
            for x in range(1,len(ans)+1):
                if(ans[x]):
                    res[i].update({x:ans[x]})
        
        
    #checking if document is same
        commonKey=[]
        for key in res[0]:
            for jey in res[1]:
                if key == jey:
                    commonKey.append(key)
        
        # with open(directory+"49.txt") as reader:
    #     f = reader.read()
    #     tokens = nltk.word_tokenize(f)
    #     print(tokens)
    #     f = re.sub(r'[^a-z1-9]+', ' ', f.lower())
    #     tokens = nltk.word_tokenize(f)
    #     print(tokens)
    #     print(process(tokens))
    #     print(nltk.pos_tag(['They'])[0])
    #     str = "thats a nice guy. basit a bad guy. There's a guy"
    #     tags = pos_tag(nltk.word_tokenize(str))
    #     print(tags)
    
        diff={}
        #calculating the positions of both words
        for x in range(len(commonKey)):
            temp1=res[0][commonKey[x]]
            temp2=res[1][commonKey[x]]
            absolVal=[]        
        #calulating difference between both words
            for y in range(len(temp1)):
                for z in range(len(temp2)): 
                    absolVal.append((abs(temp1[y]-temp2[z])))
                
            diff[commonKey[x]]=absolVal        
     
        docId=[]
        #checking if the difference is less than = to the value distance provided
        for key,value in diff.items():
            for idval in value:
                hold=len(queryTok[0])
                if(int(idval)<=hold+4*int(queryTok[2])+1):
                    docId.append(key)
        
        #appending annswers
        docId=sorted(set(docId))
        block2.delete(0,END)
        block2.insert(15,set(docId))
```

```python
#inserting files
D={}
for x in range(50):
    x=x+1
    D[x]= open(""+str(x)+".txt",encoding='utf-8').read()

concatFile=""
for x in range(50):
    x=x+1
    concatFile=concatFile+" \n"+D[x]
    
#tokenizing files
tokens=nltk.word_tokenize(concatFile)

tokens=list(set(tokens))
#removing special characters
removetable=str.maketrans("", "", "'!@#$%^&*()_=-\|][:';:,<.>/?`~")
tokens=[x.translate(removetable) for x in tokens]

#removing stopwords
stopWord=open("Stopword-List.txt").read()
stopWord=nltk.word_tokenize(stopWord)

tokens=set(tokens)-set(stopWord)
tokens=list(tokens)

tokens=sorted(tokens)
positional_ind={}

#generating positional indexes
for x in range(len(tokens)):
    positional_ind[x]={}
    for y in range(1,50):
        positional_ind[x][y]=[pos.start() for pos in re.finditer(tokens[x],D[y])]
        
        
 #generating dictionary       
dict={}
for x in range(len(tokens)):
    dict.update({tokens[x]:positional_ind[x]})
    print(positional_ind[x])

    
```
